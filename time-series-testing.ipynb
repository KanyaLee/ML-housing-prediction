```python
#copy from other files

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder, StandardScaler

df = pd.read_csv('df_cleaned.csv')

df = df.drop('name', axis=1)
encoder = LabelEncoder()
df['encoded_district'] = encoder.fit_transform(df['district'])

df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year

####
district_avg_price = df.groupby('encoded_district')['price_sqm'].mean().reset_index()
district_avg_price.columns = ['encoded_district', 'avg_price_sqm']
df = df.merge(district_avg_price, on='encoded_district', how='left')

district_avg_price_all = df.groupby('district')['price_sqm'].mean().reset_index()
district_avg_price_all.columns = ['district', 'avg_price_sqm_all']


train_df = df[df['year'] <= 2017]
test_df = df[df['year'] >= 2018]


districts = train_df['encoded_district'].unique()


for district in districts:

    district_train_df = train_df[train_df['encoded_district'] == district]
    district_test_df = test_df[test_df['encoded_district'] == district]


    X_train = district_train_df.drop(['price_sqm', 'avg_price_sqm'], axis=1)
    y_train = district_train_df['price_sqm'] - district_train_df['avg_price_sqm']
    X_test = district_test_df.drop(['price_sqm', 'avg_price_sqm'], axis=1)
    y_test = district_test_df['price_sqm']


    X_train = X_train.drop(['date', 'district', 'encoded_district'], axis=1)
    X_test = X_test.drop(['date', 'district', 'encoded_district'], axis=1)


    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)


    model = RandomForestRegressor(n_estimators=100)


    model.fit(X_train, y_train)


    y_pred = model.predict(X_test) + district_test_df['avg_price_sqm']


    district_name = encoder.inverse_transform([district])[0]


    plot_df = pd.DataFrame({'year': district_test_df['year'], 'actual': y_test, 'predicted': y_pred})


    actual_2016_2017 = train_df[(train_df['year'].isin([2016, 2017])) & (train_df['encoded_district'] == district)][['year', 'price_sqm']]
    actual_2016_2017.columns = ['year', 'actual']
    plot_df = pd.concat([plot_df, actual_2016_2017])


    plot_df = plot_df.sort_values('year')


    plt.figure()
    plt.plot(plot_df['year'], plot_df['actual'], label='Actual')
    plt.plot(plot_df[plot_df['year'] >= 2018]['year'], plot_df[plot_df['year'] >= 2018]['predicted'], label='Predicted')

    #trend line
    z = np.polyfit(plot_df[plot_df['year'] >= 2018]['year'], plot_df[plot_df['year'] >= 2018]['predicted'], 1)
    p = np.poly1d(z)
    plt.plot(plot_df[plot_df['year'] >= 2018]['year'], p(plot_df[plot_df['year'] >= 2018]['year']), label='Trend Line')

    plt.title(f'Price per Square Meter ({district_name})')
    plt.xlabel('Year')
    plt.ylabel('Price per Square Meter (THB)')
    plt.legend()
    plt.show()

    #district
    avg_price_district = district_avg_price_all[district_avg_price_all['district'] == district]['avg_price_sqm_all']
    if not avg_price_district.empty:
        avg_price_district = avg_price_district.values[0]
        print(f"The average price per square meter for the {district_name} district is: {avg_price_district:.2f} THB")
    else:
        print(f"No average price information available for the {district_name} district.")

    print(f"The predicted prices per square meter for the {district_name} district are as follows:")
    for year, price in zip(plot_df[plot_df['year'] >= 2018]['year'], plot_df[plot_df['year'] >= 2018]['predicted']):
        print(f"For the year {year}: {price:.2f} THB")
    print()
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')
```

```python
from sklearn.metrics import mean_squared_log_error
msle = mean_squared_log_error(y_test, y_pred)
print(f'Mean Squared Logarithmic Error: {msle}')
```
